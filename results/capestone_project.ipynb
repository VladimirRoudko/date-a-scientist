```python
essay_cols = ["essay0","essay1","essay2","essay3","essay4","essay5","essay6","essay7","essay8","essay9"]
# Removing the NaNs
all_essays = df[essay_cols].replace(np.nan, '', regex=True)
# Combining the essays, trimming each possible separator
all_essays = all_essays[essay_cols].apply(lambda x: ' '.join(x), axis=1)
all_essays = all_essays.apply(lambda x: x.replace(':',' '))
all_essays = all_essays.apply(lambda x: x.replace('!',' '))
all_essays = all_essays.apply(lambda x: x.replace('?',' '))
all_essays = all_essays.apply(lambda x: x.replace('*',' '))
all_essays = all_essays.apply(lambda x: x.replace(';',' '))
all_essays = all_essays.apply(lambda x: x.replace(',',' '))
all_essays = all_essays.apply(lambda x: x.replace('.',' '))
all_essays = all_essays.apply(lambda x: x.replace('(',' '))
all_essays = all_essays.apply(lambda x: x.replace(')',' '))
all_essays = all_essays.apply(lambda x: x.replace('<br',' '))
all_essays = all_essays.apply(lambda x: x.replace('/>',' '))
# get total essay length:
df["essay_len"] = all_essays.apply(lambda x: len(x))
# get list of words in each assay:
df["words"] = all_essays.apply(lambda x: x.split())
#get average word length in each assay:
df["word_count"] = df.words.apply(lambda x: len(x))
df["word_sum_len"] = df.words.apply(lambda x: sum(len(word) for word in x) )
df["avg_word_len"] = df['word_sum_len'].divide(df['word_count'], fill_value=0)

#get frequency of 'i' in the essay:
df['i_count'] = df.words.apply(lambda x: x.count('i'))
df['I_count'] = df.words.apply(lambda x: x.count('I'))
df['im_count'] = df.words.apply(lambda x: x.count("i'm"))
df['Im_count'] = df.words.apply(lambda x: x.count("I'm"))
df['i_count_total'] = df['i_count'] + df['I_count'] + df['Im_count'] + df['im_count']
df['i_freq'] = df['i_count_total'].divide(df['word_count'], fill_value=0)
df['I_freq'] = (df['I_count'] + df['Im_count']) / df['word_count']

#get frequency of 'Me' and 'me' in the essay:
df['me_count'] = df.words.apply(lambda x: x.count('me'))
df['Me_count'] = df.words.apply(lambda x: x.count('Me'))
df['me_count_total'] = df['me_count'] + df['Me_count']
df['me_freq'] = df['me_count_total'].divide(df['word_count'], fill_value=0)
df['Me_freq'] = df['Me_count'].divide(df['word_count'], fill_value=0)

```

```python
#standartize the data:
# we have total several codes:
# 'drinks_code', 'drugs_code', 'smokes_code', 'sign_code', 'sex_code', 'status_code', 'offspring_code',
# 'orientation_code', 'diet_code', "education_code', 'ethnicity_code',
# plus assay parameters:
# 'essay_len', 'avg_word_len', 'i_freq', 'I_freq','me_freq', 'Me_freq'.
from sklearn.preprocessing import StandardScaler

feature_data = df[['drinks_code', 'drugs_code', 'smokes_code', 'sign_code', 'sex_code', 
                         'status_code', 'offspring_code', 'orientation_code', 'diet_code', 
                         'education_code', 'ethnicity_code', 'income','age',
                        'essay_len', 'word_count', 'avg_word_len', 'i_freq', 'I_freq','me_freq', 'Me_freq']]


x = feature_data.values
x
scaler = StandardScaler(copy=True, with_mean=True, with_std=True)

x_scaled = scaler.fit_transform(x)
feature_data = pd.DataFrame(x_scaled, columns=feature_data.columns)
```
